```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        for page_num in range(len(pdf_reader.pages)):
            text += pdf_reader.pages[page_num].extract_text()
    return text
```

```python
#Function to calculate score based on experience
def extract_experience_score(resume_text):
    experience_score = 0
    if re.search(r"\bintern\b", resume_text, re.I):
        experience_score += 1
    if re.search(r"\bjunior\b", resume_text, re.I):
        experience_score += 2
    if re.search(r"\bsenior\b", resume_text, re.I):
        experience_score += 3
        
    years_experience_match = re.search(r"(\d+) year(s?) of experience", resume_text)
    if years_experience_match:
        years_of_experience = int(years_experience_match.group(1))
        
        if years_of_experience <= 2:
            experience_score += 1
        elif years_of_experience <= 5:
            experience_score += 2
        else:
            experience_score += 3
    return experience_score
```

```python
skills_keywords = ["Python", "Data Analysis", "Machine Learning", "SQL", "R","Pandas","NumPy","Scikit-Learn","Data Visualization","TensorFlow","Data Cleaning and Preprocessing","Data Science"]

education_keywords = [
    "University", "College", "Institute", "School", "", "Degree", "Bachelor", "Master's",
    "PhD", "Doctorate", "MBA", "Graduate", "Undergraduate", "Postgraduate", "Alumni", "Diploma", "Certificate"]


```

```python
def score_resume(resume_text):
    doc = nlp(resume_text)
    
    #initial score 
    skills_score = 0
    qualifications_score = 0
    education_score = 0
    experience_score = extract_experience_score(resume_text) 
    
      #  keep track of unique keywords
    unique_skills_keywords = set()
    
    ## Scoring based on tokens
    for token in doc:
        if token.text in skills_keywords:
            if token.text not in unique_skills_keywords:
                skills_score += 1
                unique_skills_keywords.add(token.text)
        
        if token.text in education_keywords:
            education_score += 1
            
# Calculate total score with weights
    criteria_weights = {"skills": 0.4,  "education": 0.3,"experience": 0.2}
    total_score = {
        "skills": skills_score,
       
        "education": education_score,
        "experience": experience_score,
        "total": (
            criteria_weights["skills"] * skills_score +
            
            criteria_weights["education"] * education_score +
            criteria_weights["experience"] * experience_score
        )
    }

    return total_score

resume_scores = []    

for resume in resumes:
    resume_text = resume["text"]
    total_score = score_resume(resume_text)
    
    resume_scores.append({
        "filename": resume["filename"],
        "skills": total_score["skills"],
        
        "education": total_score["education"],
        "experience": total_score["experience"],
        "total": total_score["total"]
    })
    
for score in resume_scores:
    print("Filename:", score["filename"])
    print("Skills Score:", score["skills"])
    
    print("Education Score:", score["education"])
    print("Experience Score:", score["experience"])
    print("Total Score:", score["total"])
    print("=" * 50)
    

```
